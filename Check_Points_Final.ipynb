# Checkpoint 2: Funding Type Analysis

```python
venture = master_frame['funding_round_type'] == 'venture'

angel = master_frame['funding_round_type'] == 'angel'

private_equity = master_frame['funding_round_type'] == 'private_equity'

seed = master_frame['funding_round_type'] == 'seed'
```

```python
#data frame specific to each funding type

df_venture = master_frame[venture]
df_angel = master_frame[angel]
df_private_equity = master_frame[private_equity]
df_seed = master_frame[seed]
```

```python
#Average of all the funding type

print(df_venture['raised_amount_usd'].mean())
print(df_angel['raised_amount_usd'].mean())
print(df_private_equity['raised_amount_usd'].mean())
print(df_seed['raised_amount_usd'].mean())
```
